```python
import glob
import pathlib
import os
import sys
import pandas as pd
result_folders = ['/Users/irf3irf3/Desktop/offline_workspace/data/tissue_of_origin/Griffin/Smaller_run/Nucleosome_profiling/results']
output_folder = '/Users/irf3irf3/Desktop/offline_workspace/data/tissue_of_origin/Griffin/Smaller_run/fragment_plots_median'
```

```python
GC_uncorrected_results = []
GC_corrected_results = []
for result in result_folders:
    subdirs = os.listdir(result)
    for subdir in subdirs:
        if subdir!='plots':
            if os.path.isdir(result+'/'+subdir)==False:
                continue
            bam_results = os.listdir(result+'/'+subdir)
          #  print(bam_results)
            for bam_result in bam_results:
                temp = pd.read_csv(result+'/'+subdir+'/'+bam_result, sep = '\t')
                samples =  temp['sample'].tolist()
                
                if len(set(samples))!=1:
                    print(f'multiple samples   {temp}')
                    sys.exit(1)
              
                
                temp['site_name'] = temp['site_name'].str.replace('_10000','') 
                #temp.set_index('site_name',inplace = True)
                if 'uncorrected' in bam_result:
                    GC_uncorrected_results.append(temp)
                elif 'GC_corrected' in bam_result:
                    GC_corrected_results.append(temp)
                else:
                    print('ERROR')
                    sys.exit(1)
```

```python
GC_corrected_df = pd.concat(GC_corrected_results,axis = 0)
print(GC_corrected_df.shape)
GC_corrected_df.head()
```

```python
GC_corrected_df = GC_corrected_df.drop_duplicates(subset=['sample','site_name'])
print(GC_corrected_df.shape)
GC_corrected_df.head()
```

```python
def get_label(sample_id):
    if sample_id.startswith('W') or sample_id.startswith('B'):
        return 'Bladder'
    elif sample_id.startswith('R'):
        return 'RCC'
    elif sample_id.startswith('P'):
        return 'Prostate'
    else:
        return 'Healthy'
```

```python
GC_corrected_df['status'] =  GC_corrected_df['sample'].apply(get_label)
print(GC_corrected_df.shape)
GC_corrected_df.head()
```

```python
necessary = GC_corrected_df.drop(['mean_reads_per_bp_in_normalization_window','mean_reads_per_bp_in_saved_window','mean_coverage','central_coverage','amplitude','outlier_cutoff','exclude_zero_mappability','correction','number_of_sites','amplitude', 'outlier_cutoff', 'exclude_zero_mappability', 'correction',
       'number_of_sites',  'smoothing', 'CNA_normalization','sample'], axis=1)
necessary.head()
```

```python
import matplotlib.pyplot as plt

if not os.path.exists(output_folder):
        os.makedirs(output_folder)
grouped = necessary.groupby('site_name')

for name, group in grouped:

    fig, ax = plt.subplots()

    for _, row in group.iterrows():

        color = 'peachpuff' if row['status'] == 'Healthy' else 'lightsteelblue'
        x = [int(col) for col in group.columns[0:-2]]
        y = [row[col] for col in group.columns[0:-2]]
        ax.plot(x, y, color=color)

    low_group = group[group['status'] == 'Bladder']
    high_group = group[group['status'] == 'Healthy']
    #print(low_group.iloc[:,:].head())
    low_mean = low_group.iloc[:, :-2].median(axis=0)
    high_mean = high_group.iloc[:, :-2].median(axis=0)

    ax.plot(x, low_mean, color='darkblue',  label='Bladder (n=' + str(low_group.shape[0])+")")
    ax.plot(x, high_mean, color='darkred',  label='Healthy (n=' + str(high_group.shape[0])+")")

    ax.set_title(name)
    ax.set_xlabel('Distance from site')
    ax.set_ylabel('Normalized coverage')
    ax.legend(loc='upper right', bbox_to_anchor=(1.36, 1))
    fig.savefig(f"{output_folder}/{name}.png", bbox_inches='tight',dpi=300)
    plt.close(fig)
```
