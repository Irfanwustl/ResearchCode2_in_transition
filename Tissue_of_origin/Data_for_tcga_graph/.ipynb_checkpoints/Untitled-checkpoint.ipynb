```r
library(tidyverse)
library(data.table)
library(grid)
library(gridExtra)
library(patchwork)

tcga <- read_tsv("tcga_data_summary.tsv") %>% 
  mutate(disease = factor(disease, levels = c("Bladder", "Prostate", "RCC")))
setDT(tcga)
tcga[, bin:=as.factor(rev(bin))]

print(head(tcga))

libraries <- read_tsv("libraries.tsv", show_col_types = F) %>% 
  select(-file, - library_type, - isolation_type)


print(head(libraries))

arm_z <- read_tsv("arm_z.tsv")

arm_z_wide <- arm_z %>%
  right_join(libraries, by = "library") %>%
  filter(serial == 0 | (serial == 1 & cohort == "RCC")) %>%
  filter(duplicate == 0 | (duplicate ==  1 & batch != 1)) %>%
  select(-c(count, total, prop, serial, duplicate, batch, split)) %>%
  pivot_wider(names_from = c(chr, arm), values_from = z) %>%
  mutate(cohort = factor(cohort, levels = c("Healthy", "Bladder", "Prostate", "RCC"))) %>% 
  column_to_rownames("library")

print(head(arm_z_wide))

arm_z_long <- arm_z_wide %>% 
  pivot_longer(-cohort, names_to = "arm", values_to = "z")

print(head(arm_z_long))

chromosome_order <- c('chr1_p', 'chr1_q', 'chr2_p', 'chr2_q', 'chr3_p', 'chr3_q',
                      'chr4_p', 'chr4_q', 'chr5_p', 'chr5_q', 'chr6_p', 'chr6_q',
                      'chr7_p', 'chr7_q', 'chr8_p', 'chr8_q', 'chr9_p', 'chr9_q',
                      'chr10_p', 'chr10_q', 'chr11_p', 'chr11_q', 'chr12_p', 'chr12_q',
                      'chr13_q', 'chr14_q', 'chr15_q', 'chr16_p', 'chr16_q', 'chr17_p', 
                      'chr17_q', 'chr18_p', 'chr18_q', 'chr19_p', 'chr19_q', 'chr20_p',
                      'chr20_q', 'chr21_q', 'chr22_q')

labels <- c('1p', '1q', '2p', '2q', '3p', '3q', '4p', '4q', '5p', '5q', '6p', '6q',
            '7p', '7q', '8p', '8q', '9p', '9q', '10p', '10q', '11p', '11q',
            '12p', '12q', '13q', '14q', '15q', '16p', '16q', '17p', '17q',
            '18p', '18q', '19p', '19q', '20p', '20q', '21q', '22q')

arm_z_long$arm <- factor(arm_z_long$arm, levels = rev(chromosome_order), labels = rev(labels))


print(head(arm_z_long))

arm_z_long$cohort <- factor(arm_z_long$cohort, 
                            levels = c("Healthy", "Bladder", "Prostate", "RCC"))



print(head(arm_z_long))

```
