```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier

class XGBClassifierWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, n_estimators=100, max_depth=3, learning_rate=0.1, 
                 subsample=1.0, colsample_bytree=1.0, use_label_encoder=False, 
                 eval_metric="mlogloss", objective="multi:softprob"):
        # Store all hyperparameters
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.learning_rate = learning_rate
        self.subsample = subsample
        self.colsample_bytree = colsample_bytree
        self.use_label_encoder = use_label_encoder
        self.eval_metric = eval_metric
        self.objective = objective
        
        self.model = None  # Model will be initialized later
        self.label_encoder = LabelEncoder()  # Encoder for categorical target
        self.num_classes = None  # Will be determined later

    def fit(self, X, y):
        # Convert categorical labels to numeric before fitting
        y_encoded = self.label_encoder.fit_transform(y)
        self.num_classes = len(self.label_encoder.classes_)  # Set num_class

        # Initialize the XGBClassifier with the hyperparameters
        self.model = XGBClassifier(
            n_estimators=self.n_estimators,
            max_depth=self.max_depth,
            learning_rate=self.learning_rate,
            subsample=self.subsample,
            colsample_bytree=self.colsample_bytree,
            use_label_encoder=self.use_label_encoder,
            eval_metric=self.eval_metric,
            objective=self.objective,
            num_class=self.num_classes  # ✅ Explicitly set num_class
        )

        self.model.fit(X, y_encoded)
        return self

    def predict_proba(self, X):
        return self.model.predict_proba(X)

    def predict(self, X):
        y_pred_proba = self.model.predict_proba(X)  # Get class probabilities
        y_pred_encoded = y_pred_proba.argmax(axis=1)  # Convert to class indices
        return self.label_encoder.inverse_transform(y_pred_encoded)  # Convert back to original labels


    @property
    def classes_(self):
        return self.label_encoder.classes_

```

```python
param_grid_xgb = [{
    'n_estimators': [100, 500, 1000],  # Number of boosting rounds
    'max_depth': [3, 6, 10],  # Tree depth (higher = more complex)
    'learning_rate': [0.01, 0.1, 0.2],  # Step size (lower = better generalization)
    'subsample': [0.6, 0.8, 1.0],  # Data fraction used per tree (lower = less overfitting)
    'colsample_bytree': [0.6, 0.8, 1.0],  # Features fraction used per tree (lower = less overfitting)
    'use_label_encoder': [False],  # Fixed value to avoid deprecation warning
    'eval_metric': ['mlogloss'],  # Correct metric for multi-class classification
    'objective': ['multi:softprob']  # Ensure multi-class probability output
}]

# param_grid_xgb = [{
#     'use_label_encoder': [False],  # Fixed value to avoid deprecation warning
#     'eval_metric': ['mlogloss'],  # Correct metric for multi-class classification
#     'objective': ['multi:softprob']  # Ensure multi-class probability output
# }]
```
