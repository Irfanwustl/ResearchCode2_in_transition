```python
import pandas as pd

# Load the dataframes
df1 = pd.read_excel('Supplementary Tables _102423_mineIRF.xlsx', sheet_name='irf')
df2 = pd.read_csv('OS_Survival_hypo_hyper_combined.png.txt', sep='\t') #pd.read_csv('OS_Survival_hypo_hyper_combined.png.txt', sep='\t') #pd.read_csv('PFS_Survival_hypo_hyper_combined.png.txt', sep='\t') # so or pfs
output = 'OS_out4.txt'



#numeric_cols = ['Age ', 'Baseline PSA', 'Metagene', 'ECOG PS', 'Gleason total', 'Line of Therapy in Metastatic Setting','ctDNA level']
numeric_cols = ['Age ', 'Baseline PSA', 'Metagene', 'Gleason total', 'Line of Therapy in Metastatic Setting','ctDNA level']


ordinal_col = 'Metastatic Disease Burden'
#categorical_cols = []
categorical_cols = ['Race','AR/enhancer alterations detected in cfDNA']  # Replace these with your actual column names

time_col = 'OS months (from date of sample collection)' #'PSA_progression (months)' #'OS months (from date of sample collection)'
event_col = 'OS Event' #'PSA event (1= Progression)' #'OS Event'
```

```python
import numpy as np
import pandas as pd
from lifelines import CoxPHFitter

def perform_cox_regression_with_categoricals(df, time_col, event_col, numeric_cols, categorical_cols, ordinal_col=None):
    # Create a copy to avoid warnings
    df = df.copy()
    
    # List of columns to be included
    cols_to_include = [time_col, event_col] + numeric_cols + categorical_cols
    if ordinal_col:  # Only add ordinal_col if it's not None
        cols_to_include.append(ordinal_col)
        
    df = df[cols_to_include]

    # Replace 'unknown' or 'Unknown' with NaN values
    df.replace(["unknown", "Unknown"], np.nan, inplace=True)

    # Drop rows with NaN values
    df.dropna(inplace=True)

    # Replace ordinal values
    if ordinal_col:
        ordinal_mapping = {
            ordinal_col: {
                'low': 0,
                'high': 1,
                'Low': 0,
                'High': 1
            }
        }
        df.replace(ordinal_mapping, inplace=True)

    # Update the list of categorical columns to exclude the ordinal column
    categorical_cols = [col for col in categorical_cols if col != ordinal_col]
    
    # One-hot encoding the categorical variables
    df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

    print(df_encoded.shape)
    display(df_encoded.head())

    # Fit the Cox regression model
    cph = CoxPHFitter()
    cph.fit(df_encoded, duration_col=time_col, event_col=event_col)

    # Return the model
    return cph



cph = perform_cox_regression_with_categoricals(merged_df, time_col, event_col, numeric_cols, categorical_cols, ordinal_col)
```

```python
cph.print_summary()
```

```python
import numpy as np

# Assuming cph is your fitted CoxPHFitter model

# Extract model attributes
model_info = {
    'model': str(type(cph)),
    'duration col': cph.duration_col,
    'event col': cph.event_col,
    'baseline estimation': cph.baseline_estimation_method,
    'number of observations': cph._n_examples,
    'number of events observed': cph.event_observed.sum(),  # assuming the event column is binary
    'time fit was run': cph._time_fit_was_called
}

# Convert model_info to text
model_info_text = '\n'.join([f"{k}: {v}" for k, v in model_info.items()])

# Get cph.summary as a DataFrame
cph_summary_df = cph.summary

# Convert cph_summary to text
cph_summary_text = cph_summary_df.to_string()

# Additional metrics from cph.print_summary() output:
metrics = {
    'Concordance': cph.concordance_index_,
    'Partial AIC': cph.AIC_partial_,
    'log-likelihood ratio test': cph.log_likelihood_ratio_test().test_statistic,
    '-log2(p) of ll-ratio test': -np.log2(cph.log_likelihood_ratio_test().p_value)
}

# Convert metrics to text
metrics_text = '\n'.join([f"{k}: {v}" for k, v in metrics.items()])

# Combine everything and save to a text file
with open(output, 'w') as file:
    file.write("== model_info ==\n")
    file.write(model_info_text)
    file.write("\n\n== cph.summary ==\n")
    file.write(cph_summary_text)
    file.write("\n\n== Additional Metrics ==\n")
    file.write(metrics_text)
cph_summary_df.to_csv(output+"_df.txt",sep='\t')
print("Data saved to output.txt")

```
